## Метод золотого сечения

Зададим функцию: \
$f(x) = x^2 - 5x + 2$

```python
f = lambda x: x**2 - 5*x + 2
```

Будем рассматривать функцию на отрезке [-10; 10]:

```python
x = np.linspace(-10, 10, 100)
```

```python
plt.plot(x, f(x))
pass
```

Функция для вычисления минимума методом золотого сечения:

```python
def search_min(f, a, b, accuracy):
  tau = 0.6
  n = 0

  plt.plot(np.linspace(-10, 10, 100), f(x), label=['a', 'b'])

  while (b - a > accuracy):
    plt.scatter([a, b], [f(a), f(b)], c= ['r', 'b'], alpha=0.4)
    x1 = a + (1 - tau) * (b - a)
    x2 = a + tau * (b - a)

    if (f(x1) >= f(x2)):
      a = x1
    else:
      b = x2
    min_x = (a + b) / 2
    n += 1
    
  return min_x, n
```

Зададим точность для алгоритма:

```python
accuracy = 0.01
```

Применим алгоритм к функции f(x):

```python
min_x, n = search_min(f, -10, 10, 0.01)
```

На графике точки $a$ изображены красным цветом, а точки $b$ - синим\
С каждым новым шагом точки сходятся и приближаются к минимуму,\
пока не будет достигнута заданная точность

Выведем аргумент, в котором функция принимает минимальное значение и число шагов, пройденное алгоритмом:

```python
min_x, n
```

Посчитаем минимум функции:

```python
min = f(min_x)
min
```
